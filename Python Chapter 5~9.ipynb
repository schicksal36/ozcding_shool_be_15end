## 변수에 값을 넣는 것 => 할당

"=" ": 할당연산자
"+" : 산수 연산자
+= 산술 후 바로 할당하는 문법
/=,-=,*= 모두가 가능하다

```python
# 변수를 사용하면서 가능 한 문법

print(sum)
sum2 =1 + 2
sum2 += 3
sum2 += 4
sum2 += 5
```
