## 모험가들의 레벨업 도전

### [배경 이야기]

당신은 마법사, 전사, 도적, 힐러 네 명의 모험가를 이끄는 리더입니다.

각 캐릭터는 힘, 민첩, 지능, 체력이라는 능력치를 가지고 있습니다. 이 능력치들을 이용해 다양한 도전을 해결하고 사냥을 통해 경험치를 얻어 레벨업을 해야 합니다.

<br>

### [동료 캐릭터와 능력치 정보]

마법사: 힘(10), 민첩(20), 지능(80), 체력(50)

전사: 힘(80), 민첩(30), 지능(20), 체력(100)

도적: 힘(40), 민첩(80), 지능(30), 체력(60)

힐러: 힘(20), 민첩(40), 지능(70), 체력(70)

<br>

* 변수명을 영어로 변경해서 풀어보세요 !

```python
#디셔너리 문법으로 하였습니다.
#디셔너리는 리스트인데 수정이 불가..한걸로 알고 있습니다.

wizard = {"str": 10, "ag": 20, "inte": 80, "hp": 50}
warrior = {"str": 80, "ag": 80, "inte": 20, "hp": 10}
rogue = {"str": 40, "ag": 80, "inte": 30, "hp": 60}
healer = {"str": 20, "ag": 40, "inte": 70, "hp": 70}
```

<br>

__[문제 1: 첫 번째 사냥]__

마법사는 몬스터를 사냥하기 위해 마법을 사용합니다.

마법을 사용하려면 지능이 50 이상이어야 합니다.

마법사의 지능이 50 이상인지 확인하세요.

```python
#">=" wizard가 50보다 같거나 크다
#1번 부터 3번까지 이렇게 풀었습니다.
wizard["inte"] >= 50
```

<br>

__[문제 2: 두 번째 사냥]__

전사는 강력한 공격을 하기 위해 자신의 힘을 사용합니다.

전사의 힘이 60 이상이고 체력이 80 이상인지 확인하세요.

```python
#and는 두조건 모두 True일때 쓰입니다.
warrior["str"] >= 60 and warrior["str"] >=80
```

<br>

__[문제 3: 세 번째 사냥]__

도적은 민첩함을 이용해 몬스터를 회피하고 공격합니다.

도적의 민첩이 70 이상이거나 힘이 50 이상인지 확인하세요.

```python
#or은 둘중 하나라두 True일떄 쓰입니다.
rogue["ag"] >= 70 or rogue["str"] >= 50
```

<br>

__[문제 4: 네 번째 사냥]__

힐러는 파티의 체력을 회복시키기 위해 마법을 사용합니다.

힐러의 지능이 60 이상이고, 체력이 50 이상인지 확인하세요.

```python
healer["inte"] >= 60 and healer["hp"] >= 50
```

<br>

__[문제 5: 레벨업]__

모든 캐릭터가 사냥을 성공적으로 마쳤습니다. 이제 경험치를 얻어 레벨업을 해야 합니다.


각 캐릭터의 경험치가 100 이상인지 print()를 이용해 확인해주세요.

<br>

경험치는 아래와 같습니다.

마법사: 120

전사: 150

도적: 110

힐러: 90



```python
exp = 100
wizard = 120
warrior = 150
rogue = 110
healer = 90


print(exp <= wizard)
print(exp <= warrior)
print(exp <= rogue)
print(exp <= healer)


```
